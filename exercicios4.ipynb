```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv('housing_ok.csv')
```

```python
housing.info()
```

```python
import numpy as np
condicoes = [housing.tipo == 'casa', housing.tipo == 'apartamento', housing.tipo == 'kitnet', ]
escolhas = [1, 2, 3]
housing['tipo_num'] = np.select(condicoes, escolhas) 
plt.figure(figsize=(20, 10))
plt.scatter(housing.longitude, housing.latitude, c=housing.tipo_num)
plt.colorbar()
```
