# Introdução a séries temporais

## Métodos preditivos

Previsões de séries temporais é um tópico com muitas aplicações possíveis, tais como previsão de preço de ativos, previsão do tempo, planejamento de negócios, alocação de recursos e muitos outros.

Um séries temporais é usualmente  modelada sob um processo estocástico $Y(t)$, i.e., uma sequência de variáveis aleatórias. Em uma previsão, estamos no momento $t$ e temos interesse de estimar $Y(t+1)$, usando somente as informações disponíveis até o momento $t$.

### Como validar e testar um modelo de série temporal?

Por conta da dependência temporal em uma série temporal, não podemos contar com técnicas de validação usuais. Para evitar avaliações tendenciosas, precisamos garantir que o conjunto de treinamento contem observações antecedem as observações contidados no conjunto de validação. Uma forma de superar esse problema é através do uso de "janelas de tempo". 
Se estivermos interessados em prever os próximos $n$ momentos, podemos aplicar o procedimento de validação para $1, 2, \ldots, n$ momentos a frente. Dessa forma, podemos comparar a qualidade da previsão para diferentes horizontes de tempo. 




## Modelo Naive

O modelo _Naive_ é o exemplo mais básico e "inocente" de preditor para uma série temporal. A idéia do modelo é a de que o valor previsto para o período atual $t$ seja igual ao valor do período $t-1$. Para demonstrar a aplicação do modelo _Naive_, utilizaremos os dados contidos no arquivo ```shampoo.csv```, representando as vendas mensais de shampoo.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dados/shampoo.csv')
# df.date = pd.to_datetime(df.date)
# df.set_index('date', inpla)
df.Month = pd.to_datetime(df.Month)
df.set_index('Month', inplace=True)
df.head()
````

Para avaliar a qualidade do modelo, faremos a previsão para toda a série temporal utilizando o comando ```Series.shift()```. Criaremos uma nova coluna chamada ```Sales_prediction``` como sendo o deslocamento de um período da coluna ```Sales```.

````python
df['naive_prediction'] = df.Sales.shift(1)
df.head()
````

Plotando as duas séries, é possível observar o comportamento do modelo.

````python
df.plot(figsize=(10, 6))
````

### Validação

Faremos a validação do modelo _Naive_ calculando o erro médio quadrático entre as previsões e os valores observados. Para tanto, faremos o cálculo da diferença quadrática entre a coluna ```Sales``` e a coluna ```naive_prediction```. Guardaremos a ```Series``` gerada em uma variável chamada ```errors```. Para calcular a média do erro, utilizamos o comando ```Series.mean()```.

````python
errors = (df.Sales - df.naive_prediction) ** 2
errors.mean()
````

O modelo _Naive_ é utilizado como referência para modelos mais sofisticados. A idéia é que modelos de previsão possuam erro médio quadrático menor que o modelo _Naive_. Caso contrário, temos um modelo extremamente ineficaz.

## Modelo AutoRegressivo (AR)

Em modelos de regressão múltipla, fazemos a previsão da variável de interesse utilizando a combinação linear das variáveis preditoras. Em um modelo autoregressivo, fazemos a previsão da variável de interesse utilizando a combinação linear dos valores passados da mesma variável. O termo autoregressivo indica que o processo é definido como um modelo de regressão de uma variável por ela mesma.


Assim, um modelo autoregressivo de ordem $p$ pode ser escrito como

$$y_t = c + \beta_1y_{t-1} + \beta_2y_{t-2} + \cdots + \beta_p y_{t - p} + \epsilon_t$$

onde $\epsilon_t$ é o ruído branco. Nos referimos a um modelo AR($p$) como um modelo autoregressivo de ordem $p$.

Faremos a implementação de um modelo AR(1) para o exemplo do shampoo. Para tanto, criaremos um novo ```DataFrame``` chamado ```df1``` que contém a coluna ```Sales``` e a coluna ```x1```. Onde ```x1``` representa o valor da venda com _lag_ de 1 período.

````python
df1 = df.drop('naive_prediction', axis=1)
df1['x1'] = df1.Sales.shift(1)
df1.dropna(axis=0, inplace=True)
df1.head()
````

Trataremos o modelo AR(1) como uma regressão linear simples, onde ```x1``` é a variável independente e ```Sales``` é a variável dependente. Utilizando a classe ```LinearRegression``` do pacote sklearn, podemos criar o modelo de regressão que equivale ao modelo autoregressivo AR(1).

````python
from sklearn.linear_model import LinearRegression

X = df1.x1.values
y = df1.Sales.values

reg = LinearRegression().fit(X.reshape(-1, 1), y)
reg.score(X.reshape(-1, 1), y)
````

Rodando o modelo de regressão para o AR(1) obtemos $R^2 = 0.51$. Abaixo, temos a representação visual do modelo de regressão.

````python
plt.scatter(X, y)
plt.plot(X, reg.predict(X.reshape(-1, 1)), color='red')
plt.xlabel('x1')
plt.ylabel('Sales')
````

Faremos, agora, a aplicação do modelo de regressão para AR(2). Para tanto, acrescentaremos a coluna denominada ```x2```, representando o _lag_ de 2 períodos. Nesse caso, temos um modelo de regressão linear múltipla com ```x1``` e ```x2``` como variáveis independentes e ```Sales``` como variável dependente.

````python
df1['x2'] = df1.Sales.shift(2)
df1.dropna(axis=0, inplace=True)
df1.head()
````

Criamos o novo modelo de regressão dando o nome de ```reg2```. O resultado é um modelo com ajuste superior ao anterior ($R^2 = 0.76$).

````python
X = df1.drop('Sales', axis=1).values
y = df1.Sales.values

reg2 = LinearRegression().fit(X, y)
reg2.score(X, y)
````

Para visualizar o modelo no formato da série temporal, utilizamos o código abaixo.

````python
plt.figure(figsize=(10, 6))
plt.plot(df.Sales)
plt.plot(df.index.values[3:], reg2.predict(X))
plt.legend(['Sales', 'AR(2)'])
````

### Utilizando o pacote statsmodels

Para facilitar o uso de modelos autoregressivos, utilizaremos o pacote ```statsmodels```. O modelo autoregressivo encontra-se no subpacote ```tsa.ar_model```. Para importar utilizamos o comando ```from statsmodels.tsa.ar_model import AR```. Para criar o modelo, utilizamos o comando ```AR(Series).fit(p)```, onde $p$ é o número de períodos passados (_lag_) considerados no modelo.

Faremos a criação do modelo AR(2) para comparar com o resultado gerado anteriormente. Chamaremos o modelo criado de ```ar_model2```.

````python
from statsmodels.tsa.ar_model import AR

ar_model2 = AR(df.Sales, freq=df.index.inferred_freq).fit(2)
````

Utilizando o comando ```predict()```, obtemos as previsões de cada período da série. Podemos utilizar essa informação para calcular o erro quadrático do modelo.

````python
errors = (df.Sales - ar_model2.predict()) ** 2
errors.mean()
````

Podemos, também, analisar o erro através das estatísticas descritivas do mesmo. Os erros residuais encontram-se disponíveis no modelo ```ar_model2``` no atributo ```resid```. Os valores são guardados na forma de uma ```Series``` do pandas. O que permite o uso do comando ```Series.describe()``` para cálculo das estatísticas.

````python
ar_model2.resid.describe()
````

Analisando as estatísticas descritivas do erro residual, notamos que, na média, o erro flutua ao redor do zero. O desvio padrão do erro é de 7,28 unidades, onde o maior erro para baixo foi de 121 unidades e o maior erro para cima foi de 147 unidades. 

Para analisar a distribuição dos erros residuais, criaremos o histograma dos erros utilizando o comando ```Series.hist()```.

````python
ar_model2.resid.hist()
````

Analisando a distribuição de frequência do erro, observamos que, erramos com mair frequência para baixo do que para cima.

Abaixo, temos o gráfico da séries _vs_ previsões.

````python
plt.figure(figsize=(10, 6))
plt.plot(df.Sales)
plt.plot(ar_model2.predict())
plt.legend(['Sales', 'AR(2)'])
````

Para prever um intervalo de períodos específico, faremos a configuração dos parâmetros ```start``` e ```end``` do comando ```predict()```. No exemplo, temos os valores de vendas até o período '2002-12'. Faremos a previsão de vendas para o período seguinte fazendo ```start='2003-01'``` e ```end='2003-01'```. Criaremos uma variável chamada ```forecast``` para guardar a ```Series``` resultante da previsão e plotaremos o resultado utilizando um gráfico ```plt.scatter()```.

````python
plt.figure(figsize=(10, 6))
plt.plot(df.Sales)
forecast = ar_model2.predict(start='2003-01', end='2003-01')
plt.scatter(forecast.index, forecast.values, color='red')
plt.legend(['Sales', 'Previsão'])
````

Para séries temporais, estamos interessados na previsão somente do próximo período. No exemplo, previsões para o período '2003-02' utilizariam a previsão feita no período '2003-01' como _input_, reduzindo drásticamente a qualidade da previsão.

````python
ar_model2.predict(start='2003-01-01', end='2003-02-01')
````

### Autocorrelação Total e Parcial

A autocorrelação é a medida de correlação entre uma variável e valores passados da mesma. Na autocorrelação total, a correlação entre a observação $X_t$ e a observação $X_{t - p}$ é calculada levando em consideração a dependência linear das observações intermediárias $X_{t - 1}, X_{t - 2}, \ldots, X_{t - p + 1}$. A autocorrelação parcial, por outro lado, calcula a correlação entre a observação $X_{t}$ e $X_{t - p}$ eliminando a dependência linear das observações intermediárias.


Utilizamos a autocorrelação parcial como indicador para escolher o valor $p$ para o modelo autoregressivo AR($p$). Para analisar a autocorrelação parcial utilizamos o gráfico de autocorrelação parcial, podendo ser plotado através do uso da função ```plot_pacf()``` disponível no sub pacote ```from statsmodels.graphics.tsaplots```. Passamos como parâmetro a ```Series``` ```df.Sales``` e o parâmetro ```lags=10```, para exibir somente os 10 primeiros _lags_.

````python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(df.Sales, lags=10)
plt.show()
````

O resultado obtido é um gráfico tendo como eixos $x$ e $y$ o número de _lags_ e a autocorrelação parcial, respectivamente. 

Estamos interessados no último valor de autocorrelação que não faça parte da região de baixa significância representada pela área em azul claro. O valor de autocorrelação para $p=0$ será sempre 1, uma vez que a correlação de uma variável com ela mesma sempre o será. Para $p > 0$, percebemos que $p=2$ possui é o último valor de autocorrelação parcial dentre os _lags_ com significância estatística. Sendo assim, $AR(2)$ seria a configuração mais apropriada para um modelo auto-regressivo.

## Modelo Médias Móveis (MA)

Um modelo de médias móveis de orde $q$, ou $MA(q)$ é definido como um modelo de regressão que utiliza os $q$ erros de previsão passados como input para previsão. Assim como para o modelo $AR(p)$, o modelo $MA(q)$ assume que existe relação linear entre a variável de _outpu_ e as variáveis de _input_. O modelo $MA(q)$ é representado pela equação

$$X_t = \mu + \epsilon_t + \theta_1\epsilon_{t - 1} + \theta_2\epsilon_{t-2} + \cdots + \theta_{p}\epsilon_{t - q}$$

,onde 

* $\epsilon_{t}$: ruído (erro) do período $t$
* $\mu$: média dos valores da série

Para estimar o valor de $q$, utilizamos o gráfico da autocorrelação total, selecionando o maior valor do _lag_ que pertence ao conjunto com significância estatística. Abaixo, temos o gráfico de autocorrelação para o exemplo das vendas de shampoo criado com a função ```plot_acf()``` disponível no sub pacote ```from statsmodels.graphics.tsaplots```.

````python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(df.Sales)
plt.show()
````

Para esse exemplo, escolheríamos o valor $q=3$.

Por depender dos erros de previsão, o modelo $MA(q)$ não pode ser aplicado em uma série por conta própria. Esse tipo modelo costuma ser utilizado em conjunto com modelos auto-regressivos como o $AR(p)$. A combinações dos dois modelos é conhecida como modelo $ARMA(p, q)$.

## Modelo Auto-regressivo de Médias Móveis (ARMA)

O modelo ARMA é a junção dos modelos $AR(p)$ e $MA(q)$:

* $AR(p)$: tenta explicar o efeito de momentum da série.
* $MA(q)$: tenta capturar o efeito do ruído na série. Esse efeito pode ser interpretado como eventos inesperados que afetam a observação.

O modelo criado é representado pela equação:

${\displaystyle X_{t}=c+\varepsilon _{t}+\sum _{i=1}^{p}\beta _{i}X_{t-i}+\sum _{i=1}^{q}\theta _{i}\varepsilon _{t-i}.}$

A descoberta dos valores apropriados de ${\displaystyle p}$ e ${\displaystyle q}$ no modelo $ARMA({\displaystyle p,q})$ pode ser facilitada ao mapear as funções de autocorrelação parcial para um valor estimado de ${\displaystyle p}$ e, da mesma forma, usar as funções de autocorrelação para um valor estimado de ${\displaystyle q}$. Mais informações podem ser obtidas considerando as mesmas funções para os resíduos de um modelo ajustado com uma seleção inicial de ${\displaystyle p}$ e ${\displaystyle q}$.

Para testar o uso do modelo $ARMA(p, q)$ trabalharemos com os dados climáticos contidos no arquivo ```clima_india```.

````python
from statsmodels.tsa.arima_model import ARMA

df = pd.read_csv('dados/clima_india.csv')
df.date = pd.to_datetime(df.date)
df.set_index('date', inplace=True)
df.head()
````

Criaremos o modelo para previsão da temperatura do dia, disposta na coluna ```meantemp```. Abaixo, tempos a visualização da séries temporal.

````python
import numpy as np
import matplotlib as mpl

# mpl.rc("figure", figsize=(20, 6))
plot_acf(df.meantemp, lags=100)
plot_pacf(df.meantemp, lags=20)
plt.show()
````

````python
model_arma20 = ARMA(df.meantemp, freq='D', order=(4, 1)).fit()
model_arma20.resid.describe()
# model_arma20.predict(df.meantemp)
````

## Modelo Integrado Auto-regressivo de Média Móveis (ARIMA)
ARIMA é a acrônimo que significa "Autoregressive Integrated Moving Average Model" (Modelo de médias móveis integrado de auto regressivo).

Esse acrônimo é descritivo, capturando os aspectos chave do próprio modelo. Resumidamente, estes são:

* AR: Auto regressão. Um modelo que utiliza a relação de dependência entre uma observação e um conjunto de observações passadas.
* I: Integrated. O uso de diferenças das obervações para tornar uma série estacionária.
* MA: Média móvel. Um modelo que usa a dependência entre uma observação e o erro residual das média móveis aplicadas em observações anteriores. 

Cada um desses componentes são explicitamente especificados no modelo através de parâmetros. Uma notação padrão é utilizada como ARIMA(p, d, q), onde os parâmetros são substituidos por valores inteiros para indicar o modelo ARIMA específico que está sendo utilizado. 

Os parâmetros do modelo ARIMA são definidos como:

* p: o número de observações passadas inclusas no modelo, também chamado de "ordem do lag".
* d: o número de diferenças tomadas, também chamado de "grau de diferenciação"
* q: o tamanho da janela da média móvel, também chamado de "ordem da média móvel"

O valor 0 pode ser utilizado como parâmetro, indicando que o elemento não será utilizado. Dessa forma, o modelo ARIMA pode ser configurado como um modelo ARMA, ou até mesmo um simples modelo AR, I, ou MA.

Faremos agora a importação do modelo ARIMA contido em ```statsmodels.tsa.arima_model```.

````python
from statsmodels.tsa.arima_model import ARIMA
````

Para criar o modelo, utilizaremos o comando ```ARIMA(Series, freq='frequencia_da_serie', order=(p, d, q)).fit()```. 

O parâmetro ```freq``` recebe a frequência da série. Para configurar este parâmetro, utilizaremos o comando ```Series.index.inferred_freq```. Para o parâmetro ```order```, faremos a configuração ```(3, 1, 0)```, representando um modelo autoregressivo com lag igual a três e uma diferença sucessiva para garantir a estacionaridade da série.

````python
model_arima = ARIMA(df.meantemp, freq='D', order=(5,1,2)).fit()
model_arima.resid.describe()
````

````python
plt.figure(figsize=(20, 6))
plt.plot(df.meantemp)
plt.plot(model_arima.predict(typ='levels'))
plt.legend(['meantemp', 'ARIMA(5, 1, 2)'])
````

````python
pdq_list = [(i, 1, j) for i in range(2, 10) for j in range(10)]
best = 10000
for pdq in pdq_list:
    try:
        model = ARIMA(df.meantemp, freq='D', order=(pdq)).fit()
        errors = (df.meantemp - model.predict(typ='levels')) ** 2
        if errors.mean() < best:
            print(f'pdq={pdq}, erro={errors.mean()}')
            best = errors
    except:
        pass
````
