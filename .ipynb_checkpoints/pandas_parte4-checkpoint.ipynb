# Introdução a análise de dados com Pandas
## Gráfico de dispersão (scatter)

Gráficos scatter são utilizados para comparar pares de valores com o objetivo de identificar relações de dependência. Na presença de coordenadas geográficas, o gráfico _scatter_ pode ser utilizado para visualização da distribuição de pontos.

Utilizaremos a base ```housing_ok.csv``` como exemplo.

````python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv('dados/housing_ok.csv')
housing.quartos = housing.quartos.astype('int8')
housing.banheiros = housing.banheiros.astype('int8')
housing.vagas = housing.vagas.astype('int8')
````

## Plotando coordenadas geográficas
Podemos inicialmente plotar o gráfico scatter para termos uma ideia da distribuição geográfica dos imóveis. Para isso utilizaremos a função ```plt.scatter(x, y)```.

````python
plt.scatter(housing.longitude, housing.latitude)
plt.title('Coordenadas Geográficas')
plt.xlabel('longitude')
plt.ylabel('latitude')
````

Por conta do número excessivo de coordenadas, a visualização do gráfico fica comprometida. Para melhorar a visualização podemos adicionar os comandos:
* ```plt.figure(figsize=(20, 10))```: define o tamanho da figura (no exemplo 20x10)
* ```plt.axes(aspect='equal')```: torna a escala dos dois eixos iguais

````python
plt.figure(figsize=(20, 10))
plt.axes(aspect='equal')
plt.scatter(x=housing.longitude, y=housing.latitude)
plt.title('Coordenadas Geográficas')
plt.xlabel('longitude')
plt.ylabel('latitude')
````

Para melhorar a quantidade de informação mostrada no gráfico scatter, podemos adicionar escala de cores e de forma.

No exemplo, utilizaremos a escala de cores para representar o preço dos imóveis. Para tanto, passaremos o argumento ```c=housing.preco``` (c é a inicial de _color_). Para mostrar a legenda de cores utilizamos o comando ```plt.colorbar()```.

````python
plt.figure(figsize=(20, 10))
plt.axes(aspect='equal')
plt.scatter(x=housing.longitude, y=housing.latitude, c=housing.preco)
plt.title('Coordenadas Geográficas')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar()
````

É possível observar uma concentração de imóveis mais caros no centro leste da cidade.

Caso a escala de cores não seja agradável, podemos modifica-la passando o argumento ```cmap=nome_da_escala```.

A lista completa de opções de escala encontram-se no link: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

Algumas opções de escala disponíveis são:
* ```viridis```
* ```plasma```
* ```inferno```
* ```magma```
* ```cividis```

````python
plt.figure(figsize=(20, 10))
plt.axes(aspect='equal')
plt.scatter(x=housing.longitude, y=housing.latitude, c=housing.preco, cmap='magma')
plt.title('Coordenadas Geográficas')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar()
````

Para utilzar a escala de forma, basta passarmos o argumento ```s=DataFrame.nome_da_coluna```. 

Para o exemplo, utilizaremos o parâmetro de forma para visualizar o tamanho dos imóveis através da área construida. Dividiremos a área contruida por $5$ para reduzir o tamanho dos pontos.

````python
plt.figure(figsize=(20, 10))
plt.axes(aspect='equal')
plt.scatter(x=housing.longitude, y=housing.latitude, c=housing.preco, 
            s=housing.area_construida / 5, cmap='magma')
plt.title('Coordenadas Geográficas')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar()
````

Ainda assim, por conta da quantidade de pontos, a visualização das informações não está perfeita.

Podemos passar um último argumento chamada ```alpha```. Nesse argumento, passaremos o valor de transparência dos pontos mostrados no gráfico. O valor padrão é 1.0 (100%). Para o exemplo, faremos ```alpha=0.6```.

````python
plt.figure(figsize=(20, 10))
plt.axes(aspect='equal')
plt.scatter(x=housing.longitude, y=housing.latitude, c=housing.preco, 
            s=housing.area_construida / 5, cmap='magma', 
            alpha=0.6)
plt.title('Coordenadas Geográficas')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar()
````

## Verificando relação entre atributos da tabela
O gráfico _scatter_ pode ser utilizado para verificar a existência de relação entre colunas da tabela.  No exemplo, verificaremos a relação entre a área construída e o preço do imóvel.

````python
plt.scatter(housing.area_construida, housing.preco)
plt.xlabel('Área construída')
plt.ylabel('Preço do imóvel')
plt.title('Área x Preço')
````

É possível perceber que existe uma relação forte entre a área construída e o preco dos imóveis (possivelmente linear).

Existem, porém, alguns pontos (poucos) que não obedecem o padrão geral observado. Chamamos esses pontos de _outliers_ (pontos fora da curva).
Os _outliers_ podem atrapalhar a análise de correlação e devem ser eliminados.

Para isso, acrescentaremos um filtro para eliminar valores de área contruída maiores que 1500 e preços maiores que 5 milhões. Para criar filtros com mais de uma condição, utilizamos a notação ```(condição_1) & (condição_2)```. O símbolo ```&``` é utilizado para explicitar que as duas condições devem ser respeitadas simultaneamente.

````python
filtro = (housing.area_construida <= 1500) & (housing.preco <= 5000000)
plt.scatter(x=housing[filtro].area_construida, y=housing[filtro].preco)
plt.xlabel('Área construída')
plt.ylabel('Preço do imóvel')
plt.title('Área x Preço')
````

Ainda assim, observamos um padrão linear de relação entre o preço dos imóveis e a área construída dos mesmos.
